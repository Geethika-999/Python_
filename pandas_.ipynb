Merge and Join dataframes

```python

df1 = pd.DataFrame({'student':['Ram','Vinay','Raju','Ravi'],
                    'group':['ECE','EEE','EEE','CSE']})

df2 = pd.DataFrame({'student':['Raju','Ram','Vinay','Ravi'],
                    'marks':[40,50,60,89]})
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({'group': ['ECE', 'EEE', 'CSE'],
                    'supervisor': ['Carly', 'Guido', 'Steve']})
df4
```

```python
pd.merge(df3, df4)
```

```python
df5 = pd.DataFrame({'group': ['ECE', 'ECE',
                              'EEE', 'EEE', 'CSE', 'CSE'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})
pd.merge(df1, df5)
```

```python
pd.merge(df1, df2, on='student')
```

```python
df3 = pd.DataFrame({'name': ['Ram','Vinay','Raju','Ravi'],
                    'salary': [70000, 80000, 120000, 90000]})
df1,df3
```

```python
pd.merge(df1, df3, left_on="student", right_on="name")
```
